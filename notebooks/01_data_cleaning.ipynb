### Cleaning 'County'

```python
df_county = df_state.copy()
```

```python
df_county['County'].unique()
```

```python
df_county['County'].value_counts()
```

```python
df_county['County'].isna().sum()
```

```python
df_county.loc[(df_county['Land Category'] == 'Offshore') & (df_county['County'].isna()), 'County'] = 'Offshore'
df_county.loc[(df_county['Land Class'] == 'Native American') & (df_county['County'].isna()), 'County'] = 'Native American'
```

```python
df_county['County'].isna().sum()
```

### Cleaning 'FIPS Code'

```python
df_fips = df_county.copy()
```

```python
df_fips['FIPS Code'].unique()
```

```python
df_fips['FIPS Code'].isna().sum()
```

```python
df_fips.loc[df_fips['FIPS Code'].isna(), 'FIPS Code'] = 0

# Alternatively
#df_fips['FIPS Code'] = df_fips['FIPS Code'].fillna(0)
```

```python
df_fips['FIPS Code'].isna().sum()
```

### Cleaning 'Offshore Region'

```python
df_offshoregion = df_fips.copy()
```

```python
df_offshoregion['Offshore Region'].unique()
```

```python
df_offshoregion['Offshore Region'].value_counts()
```

```python
df_offshoregion['Offshore Region'].isna().sum()
```

```python
df_offshoregion['Land Category'].value_counts()
```

```python
df_offshoregion.loc[df_offshoregion['Offshore Region'].isna() , 'Offshore Region'] = 'Onshore'
```

```python
df_offshoregion['Offshore Region'].isna().sum()
```

### Cleaning 'Commodity', 'Disposition Code', 'Disposition Description', 'Volume'

```python
df_offshoregion.info()
```

```python
df_cleaned = df_offshoregion.copy()
```

```python
df_cleaned['Volume'] = df_cleaned['Volume'].replace({',':''}, regex=True).astype(float)
```
